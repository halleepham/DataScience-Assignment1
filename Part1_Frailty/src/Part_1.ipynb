Save raw data into a .csv file and create data frame

```python
import pandas as pd
import numpy as np

project_path = 'content/drive/MyDrive/DataScience-Assignment1/Part1_Frailty'

frailty_df = pd.read_csv('/content/drive/MyDrive/DataScience-Assignment1/Part1_Frailty/data_raw/frailty_data_raw.csv')
frailty_df
```

## (a) Unit standardization

```python
# Convert height in inches to meters
frailty_df['Height'] = frailty_df['Height'] * 0.0254

# Convert weight in pounds to kilograms
frailty_df['Weight'] = frailty_df['Weight'] * 0.45359237

# Rename columns
frailty_df.rename(columns={'Height': 'Height_m', 'Weight': 'Weight_kg'}, inplace=True)
frailty_df
```

## (b) Feature Engineering

i.   BMI = Weight_kg / (Height_m ** 2)



```python
# Create and calculate 'BMI' column
frailty_df['BMI'] = (frailty_df['Weight_kg'] / (frailty_df['Height_m'] ** 2)).round(2)
frailty_df
```

ii. AgeGroup (categorical) based on Age

```python
# Create and calculate 'Age Group' column
frailty_df['Age Group'] = pd.cut(x=frailty_df['Age'],
                                 bins = [0, 30, 46, 61, np.inf],
                                 labels = ['<30', '30-45', '46-60', '>60'],
                                 right=False)
frailty_df
```

## (c) Categorical -> Numeric Encoding

i. Binary encoding: Frailty_binary
(Y->1, N->0)

```python
# Mapping dictionary for 'Frailty_binary'
frailty_map = {'Y': 1, 'N': 0}

# Binary encode 'Frailty' -> 'Frailty_binary'
frailty_df['Frailty_binary'] = frailty_df['Frailty'].map(frailty_map)
frailty_df.drop('Frailty', axis=1, inplace=True)
frailty_df
```

ii. One-hot encode AgeGroup into columns

```python
# One-hot encode AgeGroup into multiple columns, with prefix 'AgeGroup', and 0 or 1
frailty_df = pd.get_dummies(frailty_df, columns=['Age Group'], prefix='AgeGroup', dtype=int)
frailty_df
```

Saving cleaned data to data_processed folder

```python
frailty_df.to_csv('/content/drive/MyDrive/DataScience-Assignment1/Part1_Frailty/data_processed/frailty_data_clean.csv', index=False)
```

## (d) EDA & Reporting

Loading cleaned data (for reproducibility)

```python
clean_frailty_df = pd.read_csv('/content/drive/MyDrive/DataScience-Assignment1/Part1_Frailty/data_processed/frailty_data_clean.csv')
```

i. Compute summary table

```python
# Specify columns to include in summary table
# (don't use encoded columns because they represent categorical variables)
summary_columns = ['Height_m', 'Weight_kg', 'Age', 'Grip Strength', 'BMI']

# Compute summary stats for selected columns
summary_df = clean_frailty_df[summary_columns].agg(['mean', 'median', 'std']).round(2)
summary_df
```

ii. Quantify relationship between strength and frailty

```python
strength_frailty_corr = clean_frailty_df['Grip Strength'].corr(clean_frailty_df['Frailty_binary']).round(4)
print(f'Correlation between Grip Strength and Frailty: {strength_frailty_corr}')
```

Save both summary table and correlation finding to report/findings.md

```python
with open('/content/drive/MyDrive/DataScience-Assignment1/Part1_Frailty/report/findings.md', 'w') as file:
  file.write('# Data Summary Report\n\n')
  file.write(summary_df.to_markdown())
  file.write(f'\n\n## Correlation between Grip Strength and Frailty: {strength_frailty_corr}\n\n')

print("Summary table saved to reports/findings.md")
```
